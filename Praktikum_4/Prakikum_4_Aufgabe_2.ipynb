```
# b) csv laden

import pandas as pd

csv_url = "https://raw.githubusercontent.com/lvlanson/data_mining/main/Praktikum_3/data_mutal.csv"
df      = pd.read_csv(csv_url)
df.head()
```

```
df.describe()
```

```
# Scores werden über die Dichte der Punkte der nächsten Nachbarn generiert

from sklearn.neighbors import LocalOutlierFactor
import numpy as np

detector = LocalOutlierFactor(n_neighbors=20)
detector.fit(df)

factors = detector.negative_outlier_factor_
outliers_sorted = np.sort(factors)[:5]

# Scores sortieren, die am abweichendsten scores rausfiltern
print(factors)

indeces = []
for outlier in outliers_sorted:
  indeces.append(*np.where(factors==outlier)[0])

df_clean = df.drop(indeces)

df_clean.describe
```

```
# Aufgabe b) HelperFunktion für plotten
from plotly.subplots import make_subplots
from math import ceil, floor
import plotly.graph_objects as go

def plot_comparison(df_scale, df_orig, title=""):
  fig = make_subplots(rows=1, cols=2, subplot_titles=(title, "Original"))
    
  columns = len(df_scale.columns)
  for i in range(columns):
    if i == 0:
      continue
    fig.add_trace(
      go.Scatter(
          x=df_scale.iloc[:, i], 
          y=df_scale.iloc[:, 0],
          mode="markers",
          name=df_scale.columns[i]
      ),
  )

  
  for i in range(columns):
    if i == 0:
      continue
    fig.add_trace(
      go.Scatter(
          x=df_orig.iloc[:, i], 
          y=df_orig.iloc[:, 0],
          mode="markers",
          name=df_scale.columns[i]
      ),
      col=2,
      row=1
  )
  fig.show()
```

```
plot_comparison(df_clean, df)
```

```
# a) Leerzeichen als Delimiter

import pandas as pd
import requests
import re

url = "https://raw.githubusercontent.com/lvlanson/data_mining/main/Praktikum_4/Death_rate.csv"
data = requests.get(url).text
data = data.split("\n")

converted = []
for i, dat in enumerate(data):
  if i == 0:
    line = re.split("(\"\s*\")", dat)
    for i, word in enumerate(line):
      if word.replace('"', '').isspace():
        del line[i]
      else:
        line[i] = re.sub('"', "", word)
    converted.append(line)
  else:
    converted.append(re.split("\s+", dat))

for item in converted:
  del item[0]


df = pd.DataFrame(converted[1:-1], columns=converted[0])

df
```

```
from plotly.subplots import make_subplots
from math import ceil, floor
import plotly.graph_objects as go

def plot_dataframe(df, title=""):
  fig = make_subplots(rows=1, cols=1, subplot_titles=(title, "Original"))
    
  columns = len(df.columns)
  for i in range(columns):
    if i == 0:
      continue
    fig.add_trace(
      go.Scatter(
          x=df.iloc[:, i], 
          y=df.iloc[:, 0],
          mode="markers",
          name=df.columns[i]
      )
    )
  
  fig.show()
```

```
plot_dataframe(df)
```

```
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

def remove_outliers(df, n):

  detector = LocalOutlierFactor(n_neighbors=20)
  detector.fit(df)

  factors = detector.negative_outlier_factor_
  outliers_sorted = np.sort(factors)[:n]

  indices = []
  for outlier in outliers_sorted:
    indices.append(*np.where(factors==outlier)[0])

  df_outliers = df.iloc[indices, :]
  print("Removed Outliers:")
  display(df_outliers)

  df_clean = df.drop(indices)

  return df_clean
```

```
df_clean = remove_outliers(df, 10)

plot_dataframe(df_clean)
```

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly
import matplotlib


def plot_comparison(df_scale, df_orig, title=""):
  
  cols = list(matplotlib.colors.CSS4_COLORS.keys())
  fig = make_subplots(rows=1, cols=2, subplot_titles=(title, "Original"))
    
  columns = len(df_scale.columns)
  for i in range(columns):
    if i == 0:
      continue
    fig.add_trace(
      go.Scatter(
          x=df_scale.iloc[:, i], 
          y=df_scale.iloc[:, 0],
          mode="markers",
          name=df_scale.columns[i],
          marker=dict(color=cols[i])
      ),
  )

  
  for i in range(columns):
    if i == 0:
      continue
    fig.add_trace(
      go.Scatter(
          x=df_orig.iloc[:, i], 
          y=df_orig.iloc[:, 0],
          mode="markers",
          name=df_scale.columns[i],
          marker=dict(color=cols[i])
      ),
      col=2,
      row=1
  )
  fig.show()
```

```
plot_comparison(df_clean, df)
```
